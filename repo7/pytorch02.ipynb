***30*. 形態素解析結果の読み込み**

```
words = []
with open('neko.txt', 'r') as f:
  for line in f:
    node = m.parseToNode(line)
    while node:
      word = {}
      word['surface'] = node.surface
      word['base'] = node.feature.split(',')[-3]
      word['pos'] = node.feature.split(',')[0]
      words.append(word)
      node = node.next
print(len(words))
```

```
for i in range(100) :
  print(words[i])
```

確かに各形態素解析結果の表層形、基本形、品詞が抽出できた。なお、今回は各形態素解析結果を辞書型wordで作成し、それらをリストwordsにまとめて格納した。wordsの長さは20万以上あったが、このなかには文頭、文末を表すBOS, EOSも含まれているため実際の単語数はもう少し少なくなる。

***31*. 動詞**

```
verv_surface_list = []
for word in words:
  if(word['pos'] == '動詞'):
    verv_surface_list.append(word['surface'])
print(len(verv_surface_list))
```

```
for i in range(100):
  print(verv_surface_list[i])
```

たしかに動詞の表層形を抽出したリストを作成できた。作成したリストの長さはを28906であり、減っている。数が多いため100単語のみ出力した。

***32*. 動詞の原形**

```
verv_base_list = []
for word in words:
  if(word['pos'] == '動詞'):
    verv_base_list.append(word['base'])
print(len(verv_base_list))
```

```
for i in range(100):
  print(verv_surface_list[i] + '   --->   ' +verv_base_list[i])
```

動詞の原形のリストverv_base_listを作成した。リストの長さが31のverv_surface_listと同じになったためうまく作成できたと言える。分かり易さのためverv_surface_listと合わせて出力した。なお、数が多いため出力数は100にした。

***33.*「AのB」**

```
word_no = []
for idx, word in enumerate(words):
  if word['surface'] == 'の':
    left_word = words[idx - 1]['surface']
    right_word = words[idx + 1]['surface']
    word_no.append( (left_word, right_word) )
    # print(left_word + '   (の)   ' + right_word)
print(len(word_no))
```

```
for i in range(100):
  print(word_no[i][0] + '   (の)   ' + word_no[i][1])
```

'の'の左右にある単語の組をword_noリストにタプルとして格納した。その長さは10000弱だった。実際に.txtを開いて確認したところうまく抽出できていた。例えば'この'という単語には'の'が含まれているが、これは抽出されない。形態素解析では単語ごとに扱うためである。正規表現と差別化できる点の一つであろう。
